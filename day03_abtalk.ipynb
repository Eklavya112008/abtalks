```python
texts = [
    "Natural Language Processing (NLP) is a fascinating field of Artificial Intelligence.",
    "It helps computers understand human language.",
    "Machine learning models often rely on clean text data."
]
```

```python
import nltk

# Download all required resources
nltk.download("punkt")
nltk.download("punkt_tab")   # <-- new dependency
nltk.download("stopwords")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

text = "Natural Language Processing (NLP) is a fascinating field of Artificial Intelligence. It helps computers understand human language."

# Step 2: Tokenize
tokens = word_tokenize(text)

# Step 3: Remove stopwords and punctuation
stop_words = set(stopwords.words("english"))
cleaned = [w.lower() for w in tokens if w.lower() not in stop_words and w not in string.punctuation]

print("Tokens:", tokens)
print("Cleaned:", cleaned)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Suppose you have multiple texts
texts = [
    "Natural Language Processing (NLP) is a fascinating field of Artificial Intelligence.",
    "It helps computers understand human language.",
    "Machine learning models often rely on clean text data."
]

# Step 2: Tokenize
from nltk.tokenize import word_tokenize
tokenized = [word_tokenize(t) for t in texts]

# Step 3: Clean (remove stopwords + punctuation)
from nltk.corpus import stopwords
import string
stop_words = set(stopwords.words("english"))

cleaned = [
    [w.lower() for w in doc if w.lower() not in stop_words and w not in string.punctuation]
    for doc in tokenized
]

# Step 4: Bag-of-Words
# Join tokens back into strings for vectorizer
cleaned_texts = [" ".join(doc) for doc in cleaned]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(cleaned_texts)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("Bag-of-Words Matrix:\n", X.toarray())
```
