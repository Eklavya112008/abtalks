```python
sentence = input("Enter a sentence: ")

cleaned = ""
for ch in sentence:
    if ch.isalnum() or ch.isspace():
        cleaned += ch.lower()
```

```python
# Normalize whitespace
words = cleaned.split()

# Count word frequency using for + if
freq = {}
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1
```

```python
# Print results
print("\nCleaned text:", " ".join(words))
print("\nWord frequencies:")
for word in freq:
    print(word, ":", freq[word])
```
