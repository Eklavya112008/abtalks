```python
# 📦 Install required packages in Colab
!pip install --upgrade langchain langchain-community langchain-openai faiss-cpu

# 🔑 Set your OpenAI API key (replace with your actual key)
import os
os.environ["OPENAI_API_KEY"] = "sk-your_api_key_here"

# ✅ Imports
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
from langchain_core.documents import Document
import datetime

# 📝 Step 1: Create documents with metadata
docs = [
    Document(
        page_content="Neural networks are powerful for image recognition.",
        metadata={"source": "AI textbook", "category": "machine_learning", "date": "2025-01-10", "document_type": "academic"}
    ),
    Document(
        page_content="Support Vector Machines work well on smaller datasets.",
        metadata={"source": "Lecture notes", "category": "machine_learning", "date": "2024-11-05", "document_type": "academic"}
    ),
    Document(
        page_content="Blockchain ensures decentralized trust.",
        metadata={"source": "Tech blog", "category": "blockchain", "date": "2023-07-15", "document_type": "blog"}
    ),
    Document(
        page_content="Python is widely used for data science.",
        metadata={"source": "Tutorial site", "category": "programming", "date": "2026-02-01", "document_type": "tutorial"}
    ),
    Document(
        page_content="Climate change impacts agriculture significantly.",
        metadata={"source": "Research paper", "category": "environment", "date": "2022-09-20", "document_type": "academic"}
    )
]

# 🧠 Step 2: Store with embeddings
embeddings = OpenAIEmbeddings()
vectorstore = FAISS.from_documents(docs, embeddings)

# 🔍 Step 3: Filtered retrieval function
def filtered_retrieve(query, filters=None, cutoff_date=None, k=3):
    results = vectorstore.similarity_search(query, k=k)

    # Apply category filter
    if filters:
        results = [doc for doc in results if all(doc.metadata.get(k) == v for k, v in filters.items())]

    # Apply date cutoff filter
    if cutoff_date:
        cutoff = datetime.datetime.strptime(cutoff_date, "%Y-%m-%d")
        results = [
            doc for doc in results
            if "date" in doc.metadata and datetime.datetime.strptime(doc.metadata["date"], "%Y-%m-%d") >= cutoff
        ]

    return results

# 🧪 Step 4: Run queries with and without filters
queries = [
    "image recognition methods",
    "small dataset algorithms",
    "trust in decentralized systems",
    "data science languages",
    "agriculture and climate"
]

filters = [
    {"category": "machine_learning"},
    {"category": "machine_learning"},
    {"category": "blockchain"},
    {"category": "programming"},
    {"category": "environment"}
]

cutoff_date = "2024-01-01"

print("\n=== Results WITHOUT filters ===")
for q in queries:
    res = filtered_retrieve(q)
    print(f"\nQuery: {q}")
    for r in res:
        print(f"- {r.page_content} | {r.metadata}")

print("\n=== Results WITH filters ===")
for q, f in zip(queries, filters):
    res = filtered_retrieve(q, filters=f, cutoff_date=cutoff_date)
    print(f"\nQuery: {q} | Filter: {f}")
    for r in res:
        print(f"- {r.page_content} | {r.metadata}")
```
