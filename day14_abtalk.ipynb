```python
# Mock knowledge base (fictional company + scenarios)
knowledge_base = {
    "company_mission": "FictionalTech's mission is to build AI companions for education.",
    "product_launch": "FictionalTech launched EduBot in March 2026.",
    "office_location": "FictionalTech headquarters is in Pilkhuwa near Ghaziabad.",
    "custom_policy": "FictionalTech employees get 20 days of annual leave.",
    "fictional_event": "In April 2026, FictionalTech hosted the AI for Students Summit."
}

# Simple chunk retrieval (top 3 matches by keyword overlap)
def retrieve_chunks(query):
    results = []
    for key, text in knowledge_base.items():
        if any(word.lower() in text.lower() for word in query.split()):
            results.append(text)
    return results[:3]


```

```python
# Generate with RAG
def generate_with_rag(query):
    chunks = retrieve_chunks(query)
    context = "\n".join(chunks)
    prompt = f"Context:\n{context}\n\nQuestion: {query}\nAnswer:"
    return f"[Grounded Answer]\n{prompt}"

# Generate without RAG
def generate_without_rag(query):
    prompt = f"Question: {query}\nAnswer:"
    return f"[Ungrounded Answer]\n{prompt}"
```

```python
generate_with_rag("who launched Edubot in March 2026")
```
