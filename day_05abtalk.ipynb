```python
sentences = [
    "I go for a run every morning.",
    "Yoga helps me stay flexible and calm.",
    "Drinking water keeps me energized.",
    "I sleep at least 8 hours every night.",
    "Cycling is my favorite weekend activity.",
    "I avoid junk food to stay healthy.",
    "Meditation clears my mind after work.",
    "I lift weights three times a week.",
    "Walking after meals improves digestion.",
    "I track my calories using an app.",
    "Stretching prevents muscle injuries.",
    "I enjoy swimming during the summer.",
    "Green tea boosts my metabolism.",
    "I prefer home workouts over the gym.",
    "Running with friends keeps me motivated.",
    "I practice deep breathing exercises daily.",
    "I eat fruits and vegetables regularly.",
    "I limit screen time before bed.",
    "Hiking in nature refreshes my mood.",
    "I maintain a balanced diet with proteins."
]
```

```python
!pip install -q sentence-transformers
from sentence_transformers import SentenceTransformer, util
import numpy as np
```

```python
# Load pretrained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode dataset
embeddings = model.encode(sentences, convert_to_tensor=True)
```

```python
def semantic_search(query, top_k=3):
    query_embedding = model.encode(query, convert_to_tensor=True)
    scores = util.cos_sim(query_embedding, embeddings)[0]
    top_results = np.argsort(-scores.cpu())[:top_k]
    return [(sentences[i], scores[i].item()) for i in top_results]
```

```python
def keyword_search(query, top_k=3):
    query_words = query.lower().split()
    scores = []
    for i, sentence in enumerate(sentences):
        match_count = sum(word in sentence.lower() for word in query_words)
        scores.append((sentence, match_count))
    scores.sort(key=lambda x: -x[1])
    return scores[:top_k]
```

```python
query = "exercise for relaxation"

print("Semantic Search Results:")
for result in semantic_search(query):
    print(result)

print("\nKeyword Search Results:")
for result in keyword_search(query):
    print(result)
```
