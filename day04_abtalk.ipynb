```python
!pip install -q sentence-transformers
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = [
    "I love playing football.",
    "soccer is my favorite sport.",
    "The weather is sunny today.",
    "I enjoy coding in Python.",
    "Programming in Python makes me happy."
]
embeddings = model.encode(sentences, convert_to_tensor=True)

```

```python
# Compare sentence 0 and 1 (football vs soccer)
sim_football_soccer = util.cos_sim(embeddings[0], embeddings[1])

# Compare sentence 3 and 4 (coding vs programming in Python)
sim_coding_python = util.cos_sim(embeddings[3], embeddings[4])

# Compare sentence 0 and 2 (football vs sunny weather)
sim_football_weather = util.cos_sim(embeddings[0], embeddings[2])

print("Football vs Soccer:", sim_football_soccer.item())
print("Coding vs Python:", sim_coding_python.item())
print("Football vs Weather:", sim_football_weather.item())
```
